```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Sample dataset
data = [
    ("I feel sad and depressed", "negative"),
    ("I am very happy today!", "positive"),
    ("I'm worried about my exams", "negative"),
    ("Life is good, I have no complaints", "positive"),
    ("I'm feeling lonely and anxious", "negative"),
    ("Everything is awesome!", "positive"),
]

# Split data into texts and labels
texts, labels = zip(*data)

# Text preprocessing and model pipeline
stop_words = set(stopwords.words('english'))

def preprocess(text):
    words = word_tokenize(text)
    words_filtered = [word for word in words if word.lower() not in stop_words]
    return ' '.join(words_filtered)

# Preprocess texts
texts = [preprocess(text) for text in texts]

# Create a pipeline with TF-IDF vectorizer and Naive Bayes classifier
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Train the model
model.fit(texts, labels)
```

```python
def chatbot_response(user_input):
    # Preprocess the user input
    processed_input = preprocess(user_input)
    # Predict the sentiment
    sentiment = model.predict([processed_input])[0]
    # Generate a response based on the sentiment
    if sentiment == "negative":
        return "I'm sorry to hear that. It might help to talk to a professional about how you're feeling."
    else:
        return "That's great to hear! Keep up the positive vibes."

# Chatbot interaction loop
print("Hello! I'm here to help. How are you feeling today?")
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit", "bye"]:
        print("Chatbot: Take care! Goodbye!")
        break
    response = chatbot_response(user_input)
    print(f"Chatbot: {response}")
```
